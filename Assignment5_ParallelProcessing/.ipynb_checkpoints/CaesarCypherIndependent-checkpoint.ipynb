# Caesar Cypher Independent
** Note: This following cell contains some predefined functions to implement Caesar Cypher Indepedent. Please make sure you have run this cell before you run other cells in this notebook.**

```python
import time

def loadText():
    textList=[]
    condition=True
    while condition:
        text=input('Please Enter the Text You Want to Encipher: ')
        if text=='stop':
            condition=False
        else:
            textList.append(text)
    return textList

def cipher(text,key):
    import string
    stri=""
    for ch in text:
        if ch not in string.ascii_letters:
            stri+=ch
        else:
            output = chr(ord(ch) + key)
            outputNum = ord(output)
            if 64 < outputNum < 91 or 96 <outputNum < 123:
                        stri+=output
            else:
                x=chr(outputNum-26)
                stri+=x
    return stri

def CCIndependent(files,key):
    #starttime = datetime.datetime.now()
    start = time.clock()
    for text in files:
        stri=cipher(text,key)
        print(stri)
    #endtime = datetime.datetime.now()
    #print "Runtime: ",(endtime - starttime).seconds,"seconds"
    print("Runtime: ",(time.clock()-start),"seconds")
    
def loadDocuments():
    textList=[]
    condition= True
    while condition:
        filename=input('Please Enter the Text You Want to Encipher: ')
        if filename=='stop':
            condition=False
        else:
            with open(filename) as f:
                text=f.read()
                textList.append(text)
    return textList
```

## Encrpyt multiple documents
Caesar Cypher Independent notebook can encrpyt more than one documents and it will encrypt one by one. Once the function CCIndependent( ) is done, it will output the runtime of the function at the end of the output.

```python
textList=loadDocuments()
key=int(input("Please Enter Shift Key: "))
CCIndependent(textList,key)
```
