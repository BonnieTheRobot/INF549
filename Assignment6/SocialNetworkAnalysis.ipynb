# Social Network Analysis##
** Note: This following cell contains some predefined functions to implement Social Network Analysis. Please make sure you have run this cell before you run other cells in this notebook.**

```python
import networkx as nx
import itertools
import matplotlib.pyplot as plt

def loadOrganization():
    filename=raw_input('Please Enter the File Name:')
    with open(filename) as f:
        data=f.readlines()
        features=data[0].rstrip().split(',')[1:]
        values=[entry.rstrip().split(',')[1:] for entry in data[1:]]
        return features,values

def Organization_Visualization():
    features,values=loadOrganization()
    g=nx.Graph()
    weightedNodes={}
    for value in values:
        lst=[]
        for index,node in enumerate(value):
            if node=='1':
                lst.append(features[index])
        for i in itertools.combinations(lst,2):
            weightedNodes[i]=weightedNodes.get(i,0) + 1
    for nodes,weights in weightedNodes.items():
        #print nodes,weights
        g.add_edge(*nodes,weight=weights*0.4)
    edges = g.edges()
    weights = [g[u][v]['weight'] for u,v in edges]
    nx.draw(g,with_labels = True,node_color="#E69F00",width=weights,font_size=8,font_color='#0F0C82',edge_color="#A9A9A9")
    plt.show()   
```
