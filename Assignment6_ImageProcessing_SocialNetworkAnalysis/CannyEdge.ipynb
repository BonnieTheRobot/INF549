# Canny Edge Detection
** Note: This following cell contains some predefined functions to implement Canny Edge Detection. Please make sure you have run this cell before you run other cells in this notebook. **

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import feature,io,color
#import os

def loadIMG():
    filename=raw_input("Pleanse input the image filename:")
    #filename = os.path.abspath('rocks.jpg')
    img = io.imread(filename)
    if len(img.shape)!=2:
        img = color.rgb2gray(img)
    return img

def CannyEdge(img):
    plt.clf()
    edges1 = feature.canny(img, sigma=1)
    edges2 = feature.canny(img, sigma=3)
    fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),sharex=True, sharey=True)
    ax1.imshow(img, cmap=plt.cm.gray)
    ax1.axis('off')
    ax1.set_title('Original Image', fontsize=10)
    ax2.imshow(edges1, cmap=plt.cm.gray)
    ax2.axis('off')
    ax2.set_title('Canny Edge Image, $\sigma=1$', fontsize=10)
    ax3.imshow(edges2, cmap=plt.cm.gray)
    ax3.axis('off')
    ax3.set_title('Canny Edge Image, $\sigma=3$', fontsize=10)
    fig.tight_layout()
    plt.show()  
```

## Edge Detection with Different Paramters
The following cell will detect the edges of a given image. You can choose to use images provided in this repository or your own images.

```python
img=loadIMG()
CannyEdge(img)
```
